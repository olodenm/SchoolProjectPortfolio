#Задание 1. Загрузка данных

```python
import pandas as pd
import numpy as np
import random as r
```

```python
# читаем файл csv без использования первой строки в качестве заголовка
# загружаем df используя индексы колонок, которые хотим читать
# задаем название колонок согласно документации и учитываем все пропуски в df
df = pd.read_csv('/content/horse_data.csv',
                  header=None,
                  usecols=[0, 1, 3, 4, 5, 6, 10, 22 ],
                  names=['surgery?', 'age', 'rectal_temperature', 'pulse',
                         'respiratory_rate', 'temperature_of_extremities',
                         'pain', 'outcome'],
                  na_values=['?'])
df.head()
```

#Задание 2. Первичное изучение данных
Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.



```python
# Заменим значение "9" на 2, согласно документации (.names)
df['age'] = df['age'].replace(9, 2)
# смотрим значенмя в стоблце
df['age'].value_counts()
```

```python
# смотрим общую информацию о df
# (не знаю, нужно ли переводить все столбцы к одному Dtype (?))
df.info()
```

```python
# базовые статистики для непрерывных величин
basic_statistics = df[['rectal_temperature', 'pulse','respiratory_rate']].describe()

# мода для категориальных величин
moda = df[['surgery?', 'age', 'temperature_of_extremities', 'pain', 'outcome']].mode()

# Анализ распределения по категориям:
values = df[['surgery?', 'age', 'temperature_of_extremities', 'pain', 'outcome']].value_counts()

#Подсчет количества пропусков в каждом столбце
missing_values = df.isnull().sum()

# Проверка уникальных значений на предмет ошибок
unique_surgery = df['surgery?'].unique()
unique_age = df['age'].unique()
unique_temperature = df['temperature_of_extremities'].unique()
unique_pain = df['pain'].unique()
unique_outcome = df['outcome'].unique()
unique_values = (unique_surgery, unique_age, unique_temperature, unique_pain, unique_outcome)

# медиана для каждого столбца
columns = ['rectal_temperature', 'pulse', 'respiratory_rate']
medians = df[columns].median()

# выбросы
q1 = df[columns].quantile(0.25)
q3 = df[columns].quantile(0.75)
iqr = q3 - q1
outliers = (df[columns] < (q1 - 1.5 * iqr)) | (df[columns] > (q3 + 1.5 * iqr))

result = f"Basic Statistics:\n{basic_statistics}\n\nMedians:\n{medians}\n\nModa:\n{moda}\n\nValues:\n{values}\n\nMissing_values:\n{missing_values}\n\nUnique colums values:\n{unique_values}\n\nOutliers:\n{outliers.sum()}"
print(result)
```

```python
# Проведем анализ данных с учетом пропусков. Цель анализа:
#  чем закончилось лечение лошадей в (%) взрослых и молодых,
#  у которых была операция или операции не было
# разделим df на четыре новых df по столбцу 'age' и 'surgery?'
#  1) в первом старые лошади (1) прощедшие операцию (1),
#  2) во втором старые лошади(1) без операции (2),
#  3) в третьем молодые лошади (2) прощедшие операцию (1),
#  4) в четвертом молодые лошади (2) без операции (2)

total_old_horse_sur = df[(df['age'] == 1) & (df['surgery?'] == 1)]['outcome'].count()
total_old_horse_nosur = df[(df['age'] == 1) & (df['surgery?'] == 2)]['outcome'].count()
total_young_horse_sur = df[(df['age'] == 2) & (df['surgery?'] == 1)]['outcome'].count()
total_young_horse_nosur = df[(df['age'] == 2) & (df['surgery?'] == 2)]['outcome'].count()

# Подсчет процентного соотношения каждой категории внутри группы
df_old_horse_percent_sur = (df[(df['age'] == 1) & (df['surgery?'] == 1)].groupby('outcome')['surgery?'].count() / total_old_horse_sur) * 100
df_old_horse_percent_nosur = (df[(df['age'] == 1) & (df['surgery?'] == 2)].groupby('outcome')['surgery?'].count() / total_old_horse_nosur) * 100
df_young_horse_percent_sur = (df[(df['age'] == 2) & (df['surgery?'] == 1)].groupby('outcome')['surgery?'].count() / total_young_horse_sur) * 100
df_young_horse_percent_nosur = (df[(df['age'] == 2) & (df['surgery?'] == 2)].groupby('outcome')['surgery?'].count() / total_young_horse_nosur) * 100

# Разница между группами в процентах
difference_percent_sur = (df_old_horse_percent_sur - df_young_horse_percent_sur)
difference_percent_nosur = (df_old_horse_percent_nosur - df_young_horse_percent_nosur)

result = f"""
1)Старые лошади прощедшие операцию:
{df_old_horse_percent_sur}\n
Молодые лошади прощедщие операцию:
{df_young_horse_percent_sur}\n
Старые лошади без операции:
{df_old_horse_percent_nosur}\n
Молодые лошади без операции::
{df_young_horse_percent_nosur}\n
Разница в процентах между старыми и молодыми лодащьми прощедщими операцию:
{difference_percent_sur.reset_index()}\n
Разница в процентах между старыми и молодыми лодащьми без операции:
{difference_percent_nosur.reset_index()}\n
"""
print(result)
```

#Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
df_nopass = df.copy()
```

```python
# Расчет количества пропусков в каждом столбце
df_nopass.isna().sum()
```

```python
#  Заполнение пропусков для категориальных столбцов

# 1)столбец "surgery" пропуск заполнил 1. самое частое значение (276)
df_nopass['surgery?'].fillna((1), inplace=True)

# 2). temperature of extremities заполняю модой.
df_nopass['temperature_of_extremities'].fillna(df_nopass['temperature_of_extremities'].mode()[0], inplace=True)

# 3) pain заполняю модой.
df_nopass['pain'].fillna(df_nopass['pain'].mode()[0], inplace=True)

# 4) Заполним столбец самым частым значением (lived)
df_nopass['outcome'].fillna((1), inplace=True)

# ? А если мы хотим заполнить  заполнить пропуски в столбце например
#'temperature_of_extremities', чтобы понять как лучше заполнить пропуски
# можно расчитать кореляцию с другим столбцом 'rectal_temperature',
# Например, при повышенной ректальной температуре у лошади
# 'temperature_of_extremities' обычно "2.0" (Warm)
#но при этом чтобы узнать карреляцию не должно быть пропусков?
```

```python
# Заполнение пропусков в количественных столбцов

# 1) Заполнил пропуски согласно документации нормальным значением (normal temp is 37.8)
df_nopass['rectal_temperature'].fillna((37.8), inplace=True)

# 2)Согласно документации нормальный пульс для взрослой лошади (surgery == 1)
# между 30-40 упростим задачу и запишем среднее значение (37.85)
average_pulse_surgery_1 = round(df_nopass[(df_nopass['pulse'] >= 30) &
                                     (df_nopass['pulse'] <= 40) &
                                     (df_nopass['surgery?'] == 1)]['pulse'].mean(), 2)
df_nopass.loc[(df_nopass['pulse'].isna()) &
               (df_nopass['surgery?'] == 1), 'pulse'] = average_pulse_surgery_1
# оставшиеся пропуски заполнил медианой
df_nopass['pulse'].fillna(df_nopass['pulse'].median(), inplace=True)

# 3) respiratory_rate согласно документации нормальное дыхание(normal rate is 8 to 10)
# Генерируем случайные значения от 8 до 10
random_values = [r.randint(8, 10) if pd.isna(val) else val for val in df_nopass['respiratory_rate']]
# заполнил столбец "respiratory_rate" новыми значениями
df_nopass['respiratory_rate'] = random_values
```

```python
# Выбросы (outlers)
# имеется 5 выбросов в столбце pulse
df_nopass[df_nopass['pulse'] > 149]

# выбросы в столбце pulse идут по верхнему значению, если посмотреть df
# то мы можем соотнести их к тому что операция прошла не успешно,
# 80% лошадей с повышеным пульсом и прощедщими операцию скончались
# так же у них была и повышеная темпиратура
```
