По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
# Загружаем датафрейм с таблицей потребления энергии и выводим первые 5 значений
power = pd.read_csv('/content/power.csv')
power.head()
```

```python
# Заводим переменную для стран прибалтики
baltics_countries = power[(
power['country'].str.contains('Latvia|Lithuania|Estonia', case=False)
)]['country'].unique()

print(baltics_countries)
```

```python
# Фильтруем данные power по нескольким условиям и сохраняем результат в filtered_data
# Условие 1: Выбираем строки, где 'country' находится в списке стран балтики
# Условие 2: Выбираем строки, где 'category' равен 4, 12 или 21
# Условие 3: Выбираем строки, где 'year' за период с 2005 по 2010 год.
# Условие 4: Выбираем строки, где 'quantity' не принимает отрицательные значения

filtered_data = power[
    (power['country'].isin(baltics_countries)) &
    (power['category'].isin([4, 12, 21])) &
    (power['year'] >= 2005) & (power['year'] <= 2010) &
    (power['quantity'] >= 0)]

# задаем переменную для подсчета суммарного потребления стран Прибалтики
total_consumption = filtered_data['quantity'].sum()

print("Суммарное потребление стран Прибалтики:", total_consumption)
```
