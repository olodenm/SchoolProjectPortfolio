Для датафрейма log из материалов занятия создайте столбец source_type по правилам:


1.   *если источник traffic_source равен Yandex или Google, то в source_type ставится organic;*
2.   *для источников paid и email из России ставим ad;*
3. *для источников paid и email не из России ставим other;*
4. *все остальные варианты берём из traffic_source без изменений.*


```python
import pandas as pd
```

```python
# Загружаем датафрейм с таблицей visit_log и выводим первые 5 значений.
logs = pd.read_csv('/content/visit_log.csv', sep=';')
logs.head()
```

```python
# Создаем столбец source_type по первому правилу если источник traffic_source равен Yandex или Google, то в source_type ставится organic
logs.loc[
    (logs['traffic_source'].str.lower() == 'yandex') |
    (logs['traffic_source'].str.lower() == 'google'),
    'source_type'
] = 'organic'

logs.head()
```

```python
# для источников paid и email из России ставим ad;
logs.loc[(logs['region'] == 'Russia') &
    ((logs['traffic_source'] == 'paid') |
    (logs['traffic_source'] == 'email')),
    'source_type'
] = 'ad'

logs.loc[(logs.traffic_source == 'paid') |
         (logs.traffic_source == 'email')].head(10)
```

```python
# для источников paid и email не из России ставим other;
logs.loc[(logs['region'] != 'Russia') &
    ((logs['traffic_source'] == 'paid') |
    (logs['traffic_source'] == 'email')),
    'source_type'
] = 'other'

logs.loc[logs.source_type == 'other'].head(10)
```

```python
# все остальные варианты берём из traffic_source без изменений.
logs['source_type'].fillna(logs['traffic_source'], inplace=True)

logs[logs['source_type'] == logs['traffic_source']].head()
```

Используйте файл с оценками фильмов ml-latest-small/[ratings.csv](). Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.

```python
# Считываем файл 'ratings.csv' в DataFrame и выводим первые 5 значений.
ratings = pd.read_csv('/content/ratings.csv')
ratings
```

```python
# считаем количество оценок для каждого пользователя
rating_count = ratings.groupby('userId')['rating'].count().reset_index() # 671 пользователей
rating_count
```

```python
# фильтруем оставляя только тех пользователей что поставили > 100 оценок
rating_over_100 = rating_count[rating_count['rating'] > 100] # 258 пользователей
rating_over_100
```

```python
# делаем обьединение df ratings c df rating_over_100
active_user_ratings = ratings.merge(rating_over_100, on='userId', how='right')
# right join = результат содержал только те строки из "ratings", которые соответствуют активным пользователям.
active_user_ratings
```

```python
# Для каждого активного пользователя, вычислить разницу между максимальным и минимальным значением 'timestamp'
user_lifetime = active_user_ratings.groupby('userId')['timestamp'].max() - active_user_ratings.groupby('userId')['timestamp'].min()
user_lifetime.reset_index()
```

```python
# Находим среднее время активности пользователей, которые выставили более 100 оценок.
user_lifetime.mean()
```

Дана статистика услуг перевозок клиентов компании по типам.
Нужно сформировать две таблицы:


*   *таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;*
*   *аналогичную таблицу по типам выручки с указанием адреса клиента.*




К домашнему заданию №4
Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# Объедините данные из df по столбцу 'client_id'
merged_data = client_base.merge(rzd, on='client_id', how='left') \
    .merge(auto, on='client_id', how='left') \
    .merge(air, on='client_id', how='left') # используем left чтобы сохранить всех клиентов из client_base

merged_data
```

```python
# Таблица без указания адреса клиента
table_without_address = merged_data[['client_id', 'rzd_revenue', 'auto_revenue', 'air_revenue']]

table_without_address
```

```python
# Таблица с указанием адреса клиента
table_with_address = merged_data[['client_id', 'address', 'rzd_revenue', 'auto_revenue', 'air_revenue']]

table_with_address
```
